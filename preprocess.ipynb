```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python

image = cv2.imread('image.png')

if image is None:
    raise FileNotFoundError("The image file was not found.")
print(image.shape)
plt.imshow(image)
```

```python

gray_scaled_image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)


plt.imshow(gray_scaled_image ,cmap='gray')
```

```python

blurred = cv2.GaussianBlur(gray_scaled_image , (3,3) ,0)

plt.imshow(blurred , cmap='grey')
```

```python
binary = cv2.adaptiveThreshold(
    blurred,
    255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY_INV,
    15,
    1
)


plt.imshow(binary , cmap='grey')
```

```python
kernel = np.ones((3, 3), np.uint8)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original
axes[0].imshow(binary, cmap='gray')
axes[0].set_title("Original Binary")
axes[0].axis("off")

# Erosion
eroded = cv2.erode(binary, kernel, iterations=1)
axes[1].imshow(eroded, cmap='gray')
axes[1].set_title("Eroded")
axes[1].axis("off")

# Dilation
dilated = cv2.dilate(eroded, kernel, iterations=1)
axes[2].imshow(dilated, cmap='gray')
axes[2].set_title("Dilated")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
